```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading data
data=pd.read_csv("QVI_data.csv");
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
#total sales revenue
#total number of customers
#average number of transactions per customer
```

```python
#format the month and year
data['YEARMONTH'] = [''.join(x.split('-')[0:2]) for x in data.DATE]
data['YEARMONTH'] = pd.to_numeric(data['YEARMONTH'])
data['YEARMONTH']
data
```

```python
Total_Sales= data.groupby(['STORE_NBR', 'YEARMONTH']).TOT_SALES.sum() 
Total_Sales
```

```python
#total number of customers
Num_of_Cus = data.groupby(['STORE_NBR', 'YEARMONTH']).LYLTY_CARD_NBR.nunique()
Num_of_Cus
```

```python
#number of transactions per customer
trans_per_customer = data.groupby(['STORE_NBR','YEARMONTH']).TXN_ID.nunique()/data.groupby(['STORE_NBR', 'YEARMONTH']).LYLTY_CARD_NBR.nunique()
trans_per_customer
```

```python
# Average price per unit for each store and month.
average_price = data.groupby(['STORE_NBR', 'YEARMONTH']).TOT_SALES.sum()/ data.groupby(['STORE_NBR', 'YEARMONTH']).PROD_QTY.sum()
average_price
```

```python
# merge all outputs

data2 = [Total_Sales, Num_of_Cus, trans_per_customer,average_price]
measureOverTime = pd.concat(data2, join = 'outer', axis = 1)
measureOverTime
```

```python
# Rename the columns
measureOverTime.rename(columns = {'TOT_SALES': 'Total Sales', 'LYLTY_CARD_NBR': 'no_Customers', 0: 'trans_per_customer', 1: 'average_price'}, inplace = True)
measureOverTime.head(20)
```

```python
measureOverTime.isna().sum()
```

```python
#Since the dataset contains transaction details of many stores, 
# there may be some stores that did not record sales for the entire duration.
```

```python
#Pivoting the pandas.DataFrame to get all the recorded transactions for each store during the entire duration.
recorded_stores=pd.pivot_table(data, index="STORE_NBR", columns="YEARMONTH", values="TXN_ID", aggfunc="count")
recorded_stores
```

```python
 #lets find our Store numbers that do not have full observation
#sowe need to find them and remove these stores from the dataset, since they can't be control stores anyway.
```

```python
unrecorded_stores=[]
for i in recorded_stores.index:
    if recorded_stores.loc[i].isnull().any():
        unrecorded_stores.append(i)
unrecorded_stores
```

```python
unrecorded_stores
```

```python
measureOverTime.reset_index(inplace = True)
measureOverTime
```

```python
measureOverTime
```

```python
measureOverTime=measureOverTime.drop(unrecorded_stores, axis=0)
measureOverTime   #drop the unrecored stores from dataset
```

```python
pre_trial = measureOverTime[measureOverTime["YEARMONTH"].between(201807,201906, inclusive=True)]
pre_trial.head(20)
```

```python
control_stores=pre_trial[(pre_trial.STORE_NBR!=77 ) | (pre_trial.STORE_NBR!=86) |(pre_trial.STORE_NBR!=88)][["Total Sales","no_Customers","trans_per_customer","average_price"]].groupby(pre_trial.STORE_NBR).sum()
control_stores.head()  # control store except trial stores from the dataset.
```

```python
trial_stores=pre_trial[(pre_trial.STORE_NBR==77 ) | (pre_trial.STORE_NBR==86) | (pre_trial.STORE_NBR==88)][["Total Sales","no_Customers","trans_per_customer","average_price"]].groupby(pre_trial.STORE_NBR).sum()
trial_stores.head()   #trial stores from the dataset.
```

 • STORE_NBR 77:

```python
 #Getting the top five stores with the highest correlation with the trial store.

#Getting the difference between the trial store and the top five stores with the highest correlation with the trial store
```

```python
difference=control_stores.loc[control_stores.corrwith(trial_stores.loc[77], method="pearson", axis=1).nlargest(5).index]    

difference=(trial_stores.loc[77]-difference).sort_values(by="Total Sales", ascending=False)
difference["DIFFERENCE"]=difference["Total Sales"]-difference["Total Sales"].mean()
difference.sort_values(by="DIFFERENCE", ascending=False)    #   Sorting the pandas.DataFrame by the DIFFERENCE column.
```

 For STORE_NBR 77, we can see that STORE_NBR 46 would be the most suitable control store.

```python
difference=control_stores.loc[control_stores.corrwith(trial_stores.loc[86], method="pearson", axis=1).nlargest(5).index]    

difference=(trial_stores.loc[77]-difference).sort_values(by="Total Sales", ascending=False)
difference["DIFFERENCE"]=difference["Total Sales"]-difference["Total Sales"].mean()
difference.sort_values(by="DIFFERENCE", ascending=False) 
#print(" For STORE_NBR 86, we can see that STORE_NBR 57 would be the most suitable control store. ")
```

For STORE_NBR 86, we can see that STORE_NBR 57 would be the most suitable control store.

```python
difference=control_stores.loc[control_stores.corrwith(trial_stores.loc[88], method="pearson", axis=1).nlargest(5).index]    

difference=(trial_stores.loc[77]-difference).sort_values(by="Total Sales", ascending=False)
difference["DIFFERENCE"]=difference["Total Sales"]-difference["Total Sales"].mean()
difference.sort_values(by="DIFFERENCE", ascending=False) 
```

For STORE_NBR 88, we can see that STORE_NBR 165 would be the most suitable control store.

```python
#Let's compare the different control stores to each of the trial stores during the pre-trial duration to find any significant difference in the statistics.
```

```python
   #Extracting the data for each of the trial stores.
trial_stores_one=pre_trial.loc[pre_trial.STORE_NBR.isin([77])].reset_index()
trial_stores_two=pre_trial.loc[pre_trial.STORE_NBR.isin([86])].reset_index()
trial_stores_three=pre_trial.loc[pre_trial.STORE_NBR.isin([88])].reset_index()
```

```python
#   Extracting the data for each of the control stores.

control_stores_one=pre_trial.loc[pre_trial.STORE_NBR.isin([46])].reset_index()
control_stores_two=pre_trial.loc[pre_trial.STORE_NBR.isin([57])].reset_index()
control_stores_three=pre_trial.loc[pre_trial.STORE_NBR.isin([165])].reset_index()
```

```python
#concate them for better insight
stores=pd.concat([trial_stores_one, trial_stores_two, trial_stores_three, control_stores_one, control_stores_two, control_stores_three], axis=0)    #   Concatenating all the pandas.DataFrames into one.
stores
```

```python
import seaborn as sns
sns.set_style("darkgrid")
figure, axis=plt.subplots(1, 3, figsize=(20, 7))
sns.barplot(x="STORE_NBR", y="Total Sales", data=stores, ax=axis[0], palette="pastel")
axis[0].set_title("Total Sales")
sns.barplot(x="STORE_NBR", y="trans_per_customer", data=stores, ax=axis[1], palette="pastel")
axis[1].set_title("Total Customers")
sns.barplot(x="STORE_NBR", y="average_price", data=stores, ax=axis[2], palette="pastel")
axis[2].set_title("Transactions per Customer")
figure.suptitle("Comparison of the Total Sales, Total Customers, and Transactions per Customer for Each the Trial Stores and the Control Stores During the Trial Duration")
plt.show()
```

We can, once again, notice that STORE_NBR 88 slightly out-performs its control store, STORE_NBR 165, and still remains the best implementation of the trial of all the trial stores. The driver for this seems to be the purchasing customers rather than purchases per customer, as we can see that with the increase in the total customers, there's also an increase in the total sales almost identically, but the transactions per customer seem to be reasonably high for all the trial stores regardless of the total sales.

# ‣ Conclusion:
While the other trial stores performed the same as their corresponding control stores, we can see, however, that STORE_NBR 88 slightly out-performed its control store, STORE_NBR 165, in all attributes.

STORE_NBR 86 and 88 show a significant difference in terms of the total sales, but this isn't the case with STORE_NBR 77, which may be because of the way the trial was implemented for it.

Due to the maximum difference in the total sales of all the trial stores, STORE_NBR 88 remains the best implementation of the trial.

The driver for the increase in total sales seems to be the purchasing customers rather than purchases per customer — the more the customers, the higher the sales.
